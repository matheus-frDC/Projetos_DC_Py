```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

sns.set_theme(style="whitegrid")
df = pd.read_csv('/content/ecommerce_preparados.csv/MyDrive/Curso Python Data Analytics/Python para Dados/ecommerce_preparados.csv')
display(df.head())
```

1. Distribuição de Preços (Histograma):
A grande maioria dos produtos de e-commerce se concentra em uma faixa de preço mais baixa (geralmente até R$ 100 - R$ 150). Existe uma "cauda longa" para a direita, indicando que produtos premium ou kits muito caros são uma minoria.

```python
df = pd.read_csv('/content/ecommerce_preparados.csv/MyDrive/Curso Python Data Analytics/Python para Dados/ecommerce_preparados.csv')
plt.figure(figsize=(10, 6))
sns.histplot(df['Preço'].dropna(), bins=30, kde=False, color='skyblue')
plt.title('Histograma da Distribuição de Preços', fontsize=16)
plt.xlabel('Preço (R$)', fontsize=14)
plt.ylabel('Frequência', fontsize=14)
plt.tight_layout()
plt.show()
```

2. Preço vs. Nota (Dispersão):
Produtos de baixo custo têm uma densidade imensa de avaliações, variando de notas baixas a altas. Curiosamente, à medida que o preço aumenta, as notas tendem a se estabilizar em patamares mais altos, sugerindo que produtos premium costumam entregar uma qualidade mais alinhada às expectativas do cliente.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Preço', y='Nota', data=df, alpha=0.6, color='coral')
plt.title('Gráfico de Dispersão: Preço vs. Nota', fontsize=16)
plt.xlabel('Preço (R$)', fontsize=14)
plt.ylabel('Nota (Avaliação)', fontsize=14)
plt.tight_layout()
plt.show()
```

3. Correlação de Variáveis (Mapa de Calor):
Podemos observar como o volume de avaliações (N_Avaliações) possui correlação com a quantidade vendida (Qtd_Vendidos_Cod). Isso valida o "efeito manada" do e-commerce: produtos mais vendidos geram mais avaliações, e produtos com mais avaliações passam mais confiança e vendem mais.

```python
numeric_cols = ['Nota', 'N_Avaliações', 'Desconto', 'Preço', 'Qtd_Vendidos_Cod']
corr = df[numeric_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Mapa de Calor: Correlação entre Variáveis Numéricas', fontsize=16)
plt.tight_layout()
plt.show()
```

4. Volume por Público (Gráfico de Barras):
A análise de Gênero destaca para quais públicos os lojistas mais destinam os produtos. Nota-se a importância das categorias de nicho no e-commerce moderno.

```python
plt.figure(figsize=(10, 6))
genero_counts = df['Gênero'].value_counts()
sns.barplot(x=genero_counts.index, y=genero_counts.values, palette='viridis')
plt.title('Gráfico de Barras: Quantidade de Produtos por Gênero', fontsize=16)
plt.xlabel('Gênero', fontsize=14)
plt.ylabel('Quantidade de Produtos', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

5. Sazonalidade (Gráfico de Pizza):
O gráfico de Temporada revela se a plataforma vende mais focado em Primavera/Verão ou Outono/Inverno. Percebe-se que há uma parcela considerável de produtos "Atemporais" (não definidos), o que é uma excelente estratégia de vendas contínuas.

```python
plt.figure(figsize=(8, 8))
temporada_counts = df['Temporada'].value_counts()
plt.pie(temporada_counts.values, labels=temporada_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Gráfico de Pizza: Proporção de Produtos por Temporada', fontsize=16)
plt.tight_layout()
plt.show()
```

6. Distribuição de Qualidade (Densidade):
A densidade das notas apresenta um pico forte entre 4.0 e 5.0. Isso mostra que a satisfação geral do e-commerce é bastante positiva ou que os algoritmos de recomendação dão mais destaque para produtos bem avaliados.

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Nota'].dropna(), fill=True, color='purple', alpha=0.5)
plt.title('Gráfico de Densidade: Distribuição das Notas', fontsize=16)
plt.xlabel('Nota', fontsize=14)
plt.ylabel('Densidade', fontsize=14)
plt.tight_layout()
plt.show()
```

7. Desconto vs Preço (Regressão):
A linha de tendência sugere como o mercado aplica promoções: muitas vezes descontos percentuais maiores ou agressivos estão presentes em produtos de tíquete médio e alto para induzir a conversão de vendas difíceis.

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='Desconto', y='Preço', data=df, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Gráfico de Regressão: Desconto vs. Preço', fontsize=16)
plt.xlabel('Desconto (%)', fontsize=14)
plt.ylabel('Preço (R$)', fontsize=14)
plt.tight_layout()
plt.show()
```
